```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
n_points : int = 1000
x = np.linspace(0, 100, n_points)
y = np.random.rand(n_points)
```

```python
fig, ax = plt.subplots(2,1)
fig.suptitle('Diferencia entre lneas y scatter')
ax[0].plot(x, y)
ax[0].set_title('Lneas')
ax[1].scatter(x, y)
ax[1].set_title('Scatter')
```
